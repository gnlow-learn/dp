```typescript
import { document } from "jsr:@ry/jupyter-helper@0.2.0"
import { plot, rect, axisX, axisY } from "npm:@observablehq/plot@0.6.17"

{
    const seededNormal = normal.factory({ seed: 5 })
    const data = dp(1000, _ => seededNormal(0, 1), 1000, 3).p.entries()

    const color = {
        dark: "oklch(0.3 0.04 270)",
        light: "oklch(0.8 0.03 270)",
    }
    plot({
        document,
        marks: [
            rect(data, {
                x1: "0",
                x2: x => x[0]+0.02,
                y1: 0,
                y2: "1",
                fill: color.light,
                opacity: 0.3,
            }),
        ],
        x: { type: "linear", domain: [-4, 4], label: null },
        y: { grid: true, label: "p" },
        style: {
            background: color.dark,
            color: color.light,
            padding: "15px 10px 0 20px",
            fontFamily: "Pretendard Variable",
        },
        width: 500,
        height: 300,
    })
}
```
